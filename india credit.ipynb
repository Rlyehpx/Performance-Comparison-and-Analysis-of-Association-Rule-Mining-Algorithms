```python
import pandas as pd

file_paths = [
    'Desktop/Homework/data mining/as2/DATASET/Credit card transactions - India - Simple.csv',
    'Desktop/Homework/data mining/as2/DATASET/Amazon Customer Behavior Survey.csv.xls',
    'Desktop/Homework/data mining/as2/DATASET/Video Games Sales.csv',
    'Desktop/Homework/data mining/as2/DATASET/Video_Games_Sales_as_at_22_Dec_2016.csv'
]

for file_path in file_paths:

    data = pd.read_csv(file_path)
    
    unique_items = set()
    for column in data.columns:
        unique_items.update(data[column].unique())
    
    unique_items.discard(0) 
    unique_item_count = len(unique_items)
    
    print(f"file '{file_path}' : {unique_item_count}")
```
